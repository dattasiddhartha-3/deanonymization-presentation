```python
import pandas as pd
import random
import re
from collections import Counter
from nltk.corpus import stopwords
import networkx as nx
import matplotlib.pyplot as plt
```

```python
summ_names = [ '    < ENTER STUDENT NAMES AS A LIST>      ' ]
summ_texts = [ '  <ENTER SUMMATIVE TEXTS AS LIST> ' ]
```

```python
summatives = pd.DataFrame()
summatives['name'] = summ_names
summatives['text'] = summ_texts
summatives.head()
```

Use NLP to create "attributes" (we do not want to give the public access to the whole raw text)
* Example: 100 most frequent words used (excluding stop words)

```python
def text_feature(sentence):
    words = sentence.replace('.', ' ').replace(',', ' ').split(" ")
    words = [word.lower() for word in words if word not in stopwords.words("english") and word != '']
    return Counter(words).most_common(100)

words_freq = [list(dict(text_feature(summ_texts[i])).keys()) for i in range(len(summ_texts))]
```

```python
tokens = pd.DataFrame(words_freq).T
tokens.columns = ['topword_'+str(i) for i in tokens.columns]
tokens
```

```python
summ_pre = pd.merge(summatives, tokens, left_index=True, right_index=True)
```
